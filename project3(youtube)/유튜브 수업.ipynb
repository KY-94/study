```python

#문자열 슬라이싱

jumin = "990120-1234567"

print("성별: " + jumin[7]) # []대괄호를 쓰며 처음위치는 0부터 시작한다 , 7번째 정보를 뽑는다

print("연: " + jumin[0:2]) # 0번째부터 2번째직전 까지 뽑는다

print("월: " + jumin[2:4]) # 2번째부터 4번째직전 까지 뽑는다

print("일: " + jumin[4:6]) # 4번째부터 5번째직전 까지 뽑는다

print("생년월일:" + jumin[0:6]) # 0부터 6직전까지
print("생년월일:" + jumin[:6]) #0을 생략하여 적을수도 있음 , 처음부터 6직전까지
print("뒤 7자리:" + jumin[7:14])
print("뒤 7자리:" + jumin[7:]) #마지막은 생략하여 적을수도 있음, 7부터 끝까지

print("뒤 7자리(뒤에서부터):" + jumin[-7:]) # 맨뒷자리는 -1번쨰가 된다, 따라서 -7번째는 (1)이며, 끝-1번째는(7) 이다
```

```python

#딕션어리(사전) = 키와 , 값

#{} 안에 키:값 , 키:값 으로 작성한다
cabinet = {3:"유재석", 100:"김태호"} #3번키는 유재석이 쓰고있고, 100번키는 김태호가 쓰고있다




#사전의 값출력은 변수[키값],변수.get(키값) 사용한다

print(cabinet[3])       # 값출력하기 방법1
print(cabinet.get(3))   # 값출력하기 방법2

#print(cabinet[5])      # 없는키는 오류가나며 출력이 안됨
print(cabinet.get(5))   # get의 경우 없는키라도 None을 출력하고 정상작동한다
print(cabinet.get(5),"사용가능") # get의 경우 없는키라도 ,"사용가능"의 값을 출력하고 정상작동한다




#사전안에 값이 키값이 있는지 확인

print(3 in cabinet) # 값이 있는지 없는지 확인 키 in 변수 True로 나옴
print(5 in cabinet) # 값이 있는지 없는지 확인 키 in 변수 False로 나옴





#키와 값을 문자열로도 할수 있음

cabinet2 = {"A-1":"유재석", "B-1":"김태호"}
print(cabinet2["A-1"])
print(cabinet2["B-1"])




#캐비넷에 새로운 값 추가하기

cabinet2["C-1"] = "조세호" #c-1이라는 키에 조세호를 업데이트함 이미 c-1의 값이 있는 경우 원래이름에서 조세호로 업데이트되고 없었다면 그냥 추가만 된다
print(cabinet2)

#캐비넷에 값 제거하기

del cabinet2["A-1"]
print(cabinet2)

#키들만 출력
print(cabinet2.keys())
#값들만 출력
print(cabinet2.values())
#쌍으로 모두 출력
print(cabinet2.items())


#값 모두 제거
cabinet2.clear()
print(cabinet2)
```

```python
#세트 (집합set)

#중복이 안되며 순서가 없다. 변수 = {1,2,3,4} 값만 나열하면 됨
my_set = {1,2,3,3,3}
print(my_set) # 중복을 허용하지 않기 때문에 1,2,3만 출력된다

java = {"유재석", "김태호", "양세형"}
python = set(["유재석", "박명수"]) # 리스트로만들어주고나서 set로 감싼다

# java와 python을 모두 할 수 있는 개발자(교집합)
# 출력되는 순서는 보장되지 않는다
print(java & python)
print(java.intersection(python))

# java를 할수 있거나, python을 할 수 있는 개발자(합집합)
# 출력되는 순서는 보장되지 않는다
print(java | python)
print(java.union(python))

# java 할수있지만 python은 할줄 모르는 개발자(차집합)
# 출력되는 순서는 보장되지 않는다
print(java - python)
print(java.difference(python))

# 세트에 값 추가하기
# python 할 수 있는 사람이 늘어남
python.add("김태호")
print(python)

# 세트에 값 제거하기
# java를 하도 안써서까먹음
java.remove("김태호")
print(java)
```

```python
#자료 구조의 변경

menu = {"커피","우유", "주스"} #집합(세트)으로 만듬
print(menu, type(menu)) #집합(세트)로 되어있다

#리스트로 변경 []
menu = list(menu)
print(menu, type(menu)) # 리스트로 감싸면서 리스트로 바뀜

#튜플로 변경 ()
menu = tuple(menu) # 리스트를 튜플로 변경
print(menu, type(menu))

#다시 세트로 변경 {}
menu = set(menu) 
print(menu, type(menu))
```

```python

#컨티뉴와 브레이크

# 책을 읽으라고 시키는데 결석한 사람은 빼고 불러야함, 책을 안가져온 학생이 잇다면 바로 끝내야함 

absent = [2, 5] # 결석
no_book = [7] # 책을 깜빡함

for student in range(1, 11): #출석번호 1~10까지
    
    if student in absent: # 출석번호에 결석번호가 안에 있다면
        continue          # 밑에 뭐가있든 스킵하고 다음 회차로 넘어가라
    
    elif student in no_book:
        print("오늘 수업 여기 까지.{0}는 교무실로 따라와".format(student))
        break             # 밑에 뭐가 있든 반복문 바로 정지   
    
    print("{0}, 책을 읽어봐".format(student))


```

```python
#함수

# 기본
# def 변수():
#     명령실행문


def open_account(): #계좌생성
    print("새로운 계좌가생성되었습니다")

def deposit(balance, money): #입금시
    print("입금이 완료되었습니다. 잔액은 {0}원입니다.".format(balance + money))
    return balance + money # 잔액+입금액을 함수를 사용한 변수에 저장시킨다

def withdraw(balance, money): #출금시
    if balance >= money:
        print("출금되었습니다. 잔액은{0}원입니다.".format(balance - money))        
        return balance - money # 잔액-출금액을 함수를 사용한 변수에 저장시킨다     
    else:
        print("출금이 불가능합니다. 잔액은{0}원입니다.".format(balance))
        return balance

def withdraw_night(balance, money):
    
    commission = 100
    
    if balance >= money + 100:
        print("출금되었습니다. 수수료는{0}원, 잔액은{1}원입니다.".format(commission, balance - money - commission))        
        return commission, balance - money - commission   #리턴으로 변수를 두개 줄수도 있다, 튜플값으로 저장되어 진다   
    else:
        print("출금이 불가능합니다. 수수료는{0}원, 잔액은{1}원입니다.".format(commission, balance))
        return balance
    

balance = 0

balance = deposit(balance,1000) #입금
print(balance)

balance = withdraw(balance,10000) #출금불가
print(balance)

balance = withdraw(balance,100) #출금
print(balance)

balance = withdraw_night(balance, 1000) #밤 출금불가 
print(balance)

balance = withdraw_night(balance, 100) #밤 출금
print(balance)

```

```python
#가변인자를 이용한 함수 호출(서로다른 값의 개수를 입력하게 된다면 ?)
def profile(name, age, lang1,lang2,lang3,lang4,lang5):
    print("이름 : {0}\t나이: {1}\t".format(name,age), end=" ") #end ""빈칸을 적용하면 프린트문을 출력하고 한칸 띄우고 한줄로 이어써진다
    print(lang1,lang2,lang3,lang4,lang5)

profile("유재석", 20, "python", "java", "C", "C++", "C#") #5개까지는 잘나온다, 만약 할줄아는 언어가 늘면 함수의 정의를 새로해주어야함 > 귀찮다

profile("김태호", 25, "kotlin", "swift", "", "", "")#만약 2개면 함수호출시 ""으로 입력시켜 불러야함 > 귀찮다

#가변인자를 활용(*를 붙이면된다)

def profile(name, age, *language):
    print("이름 : {0}\t나이: {1}\t".format(name,age), end=" ") #end ""빈칸을 적용하면 프린트문을 출력하고 한칸 띄우고 한줄로 이어써진다
    for lang in language:
        print(lang, end=" ") #반복문으로 프린트를 하는데줄바꿈을 안하고 이어서 하도록 사용해봄
    print() #한칸띄워주는 용도

profile("유재석", 20, "python", "java", "C", "C++", "C#", "JV") #6 개여도 문제없이 출력
profile("김태호", 25, "kotlin", "swift" )# 2개여도 따로 입력값없이 출력됨


```

```python
#지역변수 = 함수내에서만 사용하고, 함수호출되면 변수가 생겼다가 호출이끝나면 없어지는것

#전역변수 = 프로그램내에서 언제 어디서든 부를수 있는 변수

gun = 10

def checkpoint(soldiers): #경계근무 나가는 사람수
    
    global gun      # 전역변수인 건을 쓰도록 정의한다 (전역변수값 사용)
    #gun = 20       # 함수내에서 사용하도록 gun을 정의한다 (쓰고나면 사라지는 지역 변수)
    
    gun = gun - soldiers
    print("[함수 내] 남은 총: {0}".format(gun))

#---------------실행 문 -----------------#
print("전체 총: {0}".format(gun)) # 전체총은 10 출력이된다 (전역 변수로 출력됨)

checkpoint(2)                    # 함수를 호출해서 gun에서 gun을 빼는데 gun으로 함수내에서 정의되지 않아 사용할수 없다고 오류뜬다 함수내 gun을 정의해줘야함

print("남은 총: {0}".format(gun)) # 함수를통해 마이너스 됬고 그게  글로벌 건에 저장되어 8이 되었다 지역변수로 쓸경우는 10으로그대로 출력된다


#Tip 함수내에서 사용할때 대부분은 그저 지역변수를 활용한다 전역변수를 사용할경우 코드문이복잡해지기때문
#보통 다음의 방법을 많이 사용


def checkpoint_ret(gun, soldiers): #건과,병사수를 따로함수에서 인자로 받고
    gun = gun - soldiers
    print("[함수 내] 남은 총: {0}".format(gun))
    return gun # 함수의 값을 gun에 저장 시킨다 (그러면 호출된함수자체가 숫자가 된것과같음)

#---------------실행 문 -----------------#
print("전체 총: {0}".format(gun)) 

gun = checkpoint_ret(gun , 2) #함수명령문은 리턴 gun 까지 되었으니 함수호출자체는 숫자로 6이 되어있을것이다 이를 다시금 지역변수로 정의 해준다

print("남은 총: {0}".format(gun))
```

```python
#다양한 출력포맷

#빈 자리는빈공간으로 두고, 오른쪽 정렬을 하되, 총10자리 공간을 확보
print("{0: >10}".format(500)) #중괄호로 설정입력, 0서부터 10자리까지 확보를하고 >오른쪽정렬을 하고 빈공간으 빈칸으로 한다(스페이스로 띄워줌)

#빈 자리는빈공간으로 두고, 오른쪽 정렬을 하되, 총10자리 공간을 확보, 양수일땐 +로 표시, 음수일땐 -로 표시
print("{0: >10}".format(500)) 
print("{0: >+10}".format(-500)) #+값의 의미는 +그자체가 아니라 양수음수를 표시한다는 뜻이다

#왼쪽정렬을하고, 빈칸으로 _로 채움
print("{0:_<+10}".format(500))

#3자리마다 콤마를 찍어주기
print("{0:,}".format(500000000000))

#3자리마다 콤마를 찍고, + - 처리
print("{0:+,}".format(500000000000))
print("{0:+,}".format(-500000000000))

#3자리마다 콤마를 찍고, 부호붙이고, 30자릿수 확보, 빈자리는 ^로 채우기
print("{0:^<+30,}".format(5000000000000))
print("{0:^<+30,}".format(5000000000000))

#소수점 출력
print("{0:f}".format(5/3))
#소수점 특정 자리수까지만 출력 (두번쨰짜리까지 표시, 세벉쨰자리에서 반올림)
print("{0:.2f}".format(5/3))



```

```python
#Pickle

#프로그램에서 사용하는 데이터를 파일형태로 저장해서, 다른사람이그파일을 열고, 읽어서 코드에서 또 쓸수있는것
import pickle
profile_file = open("profile.pickle", "wb") # 파일명 profile.pickle에 저장, 피클의 경우 바이너리형식을 사용함으로 wb로 열어주어야한다 인코딩은 필요없다
save = {"이름":"박명수", "나이":30,"취미":["축구", "골프", "코딩"]} #저장될내용을 딕션어리로 정하고, 취미의경우 리스트로 저장하며, 이값을들 변수 save에 저장시킨다
print(save)
pickle.dump(save, profile_file) #피클에 저장되어지기 위해서는 dump를 해야하며, save에 있는 정보를 profile_file변수인 생성된파일에 저장한다
profile_file.close()            #열면 닫기

profile_file = open("profile.pickle", "rb") # 파일명 profile.pickle에 있는것을 읽기형식으로, 피클의 경우 바이너리형식을 사용함으로 rb로 열어주어야한다 인코딩은 필요없다
read = pickle.load(profile_file)            # 피클의 load함수를 호출해서 열린 profile_file 내용을 read에 저장시킨다
print(read)
profile_file.close()            #열면 닫기
```

```python
# with 

# 파일을 닫을 필요없이 빠르게 작업시켜줌
import pickle

with open("profile.pickle", "rb") as read: #프로필파일.픽클 을 rb 형태로 읽는데 그걸 as read에 저장
    print(pickle.load(read))

# with문을 활용한 일반 파일 

with open("study.txt", "w", encoding="utf8") as reading:
    reading.write("파이썬 열심히 하는중")

with open("study.txt", "r", encoding="utf8") as reading:
    print(reading.read())    
```
