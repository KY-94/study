```python
#문자열 처리함수
python = "Python is Amazing"
print(python.lower()) # 모든 문장 다 소문자로처리
print(python.upper()) # 모든 문장 다 대문자로처리
print(python[0].isupper()) #변수의 0번째가 대문자인가?
print(len(python)) #변수의 길이를 말해준다
print(python.replace("Python", "Java")) # 파이썬이라는 변수를 자바라는 걸로 바꿔준다 *비파괴적 방법

index = python.index("n") #앞에서부터 순차적으로 n이라는 글자가 몇번째있는지 확인한다 (1번쨰 n찾기)
print(index)
index = python.index("n", index + 1) #n을 찾는데 index+1 번째부터 찾는다 (2번째 n찾기) 
print(index)

print(python.find("n")) #find로 n을 찾는다 (있다면 몇번째 위치인지 나온다)
print(python.find("Java")) #find로 java를 찾는다(없다면 -1의 값이 나온다)
#print(python.index("java")) #index에서는 원하는 값이 없을때는 오류를 내며 출력이 안됨

print(python.count("n")) # python이라는 변수에 n이 몇개 있는지 세준다


```

```python
#리스트
#지하철 칸별로 10명, 20명, 30명
# subway1 = 10
# subway2 = 20
# subway3 = 30

subway = [10,20,30]
print(subway)

subway = ["유재석", "조세호", "박명수"]
print(subway)

#조세호씨가 몇번째 칸에 타고 있는가?
print(subway.index("조세호")) #0번 1번 2번중 1번=2번칸에 타고있다

#하하씨가 다음 정류장에서 다음칸에 탔다
subway.append("하하")#맨끝에 어팬드 추가한다
print(subway) 

#정형돈씨를 유재석 / 조세호 사이에 태워봄
subway.insert(1,"정형돈") #1번째 위치에 삽입되어진다
print(subway) 

#뒤에서부터 한명씩 빼겠음
print(subway.pop()) #하하를 빼고,
print(subway) #남은 사람은 4명

#같은 이름의 사람이 몇명있는지확인
subway.append("유재석")
print(subway) #유재석을 추가해서 0번쨰 4번쨰가 유재석이다
print(subway.count("유재석")) #유재석을 센다 2명

# 정렬도 가능
num_list = [5,2,4,3,1]
num_list.sort() #순서대로 정렬한다
print(num_list)
num_list.reverse() #역순으로 정렬한다
print(num_list)
num_list.clear() #리스트를 비운다
print(num_list)

#리스트는 자료형에 구애받지 않고 섞어 사용가능하다
mix_list = ["조세호",20,True]

#리스트를 합칠 수도있다 (리스트 확장)
num_list = [5,2,4,3,1]
mix_list = ["조세호",20,True]
num_list.extend(mix_list)
print(num_list)
```

```python
#딕션어리(사전) -키와 , 값
cabinet = {3:"유재석", 100:"김태호"} #3번키는 유재석이 쓰고있고, 100번키는 김태호가 쓰고있다
print(cabinet[3]) #값출력하기 방법1
print(cabinet.get(3)) #값출력하기 방법2

#print(cabinet[5]) #없는키는 오류가나며 출력이 안됨
print(cabinet.get(5)) #get의 경우 없는키라도 None을 출력하고 정상작동한다
print(cabinet.get(5),"사용가능") #get의 경우 없는키라도 None을 출력하고 정상작동한다

print(3 in cabinet) # 값이 있는지 없는지 확인 키 in 변수 True로 나옴
print(5 in cabinet) # 값이 있는지 없는지 확인 키 in 변수 False로 나옴

#키와 값을 문자열로도 할수 있음
cabinet2 = {"A-1":"유재석", "B-1":"김태호"}
print(cabinet2["A-1"])
print(cabinet2["B-1"])

#캐비넷에 새로운 값 추가하기
cabinet2["C-1"] = "조세호" #c-1이라는 키에 조세호를 업데이트함 이미 c-1의 값이 있는 경우 원래이름에서 조세호로 업데이트되고 없었다면 그냥 추가만 된다
print(cabinet2)

#캐비넷에 값 제거하기
del cabinet2["A-1"]
print(cabinet2)

#키들만 출력
print(cabinet2.keys())
#값들만 출력
print(cabinet2.values())
#쌍으로 모두 출력
print(cabinet2.items())


#값 모두 제거
cabinet2.clear()
print(cabinet2)
```

```python
#튜플
#목록에 추가를 할수는 없고 변경 불가능 하지만 속도가 빠르다
menu = ("돈까스", "치즈가스")
print(menu[0])
print(menu[1])

# menu.add("생선까스") 튜플은 추가기능이 불가능하다
#튜플의 구체적 활용
#기존
name = "김종국"
age = 20
hobby = "코딩"
print(name,age,hobby)
#튜플사용
(name, age, hobby) = ("김종국", 20, "코딩") #한번에 선언가능
print(name, age, hobby)
```

```python
#세트 (집합set)
#중복이 안되며 순서가 없다.
my_set = {1,2,3,3,3}
print(my_set) #중복을 허용하지 않기 때문에 1,2,3만 출력된다

java = {"유재석", "김태호", "양세형"}
python = set(["유재석", "박명수"]) #리스트로만들어주고나서 set로 감싼다

# java와 python을 모두 할 수 있는 개발자(교집합) #출력되는 순서는 보장되지 않는다
print(java & python)
print(java.intersection(python))

# java를 할수 있거나, python을 할 수 있는 개발자(합집합) #출력되는 순서는 보장되지 않는다
print(java | python)
print(java.union(python))

# java 할수있지만 python은 할줄 모르는 개발자(차집합)
print(java - python)
print(java.difference(python))

#python 할 수 있는 사람이 늘어남
python.add("김태호")
print(python)

# java를 하도 안써서까먹음
java.remove("김태호")
print(java)
```

```python
#자료 구조의 변경
menu = {"커피","우유", "주스"}
print(menu, type(menu))

#리스트로 변경 []
menu = list(menu)
print(menu, type(menu))

#튜플로 변경 ()
menu = tuple(menu)
print(menu, type(menu))

#다시 세트로 변경 {}
menu = set(menu)
print(menu, type(menu))
```
