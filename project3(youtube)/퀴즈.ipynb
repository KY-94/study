```python

# 퀴즈3

# quiz : 사이트별로 비밀번호를 만들어 주는 프로그램

# 예: http://naver.com
# 규칙1 : http:// 부분은 제외 = ex) naver.com
# 규칙2 : 처음만나는점 . 이후 부분은 제외 = ex) naver
# 규칙3 : 남은 글자중 처음 세자리 + 글자 개수 + 글자내 'e' 개수 + "!" 로 구성

url = input("주소를 입력하세요: ") #주소를 받아온다
url = url.replace("http://" , "") # 규칙1
url = url[:url.index(".")] # 규칙 2     .의 위치 직전까지 슬라이싱
password = url[0:3] + str(len(url)) + str(url.count("e")) + "!" #규칙 3
print("비밀번호:", password)
```

```python
# 퀴즈4

# 학교이벤트 코딩대회
# 댓글작성자들중에 추첨을 통해 1명 치킨, 3명 커피

# 조건1 : 댓글은 20명이 작성, 아이디는 1~20 이라고 가정
# 조건2 : 무작위로 추첨하되 중복 불가
# 조건3 : random 모듈의 shuffle 과 sample 사용

# (출력 예시)
# == 당첨자 발표 ==
# 치킨 당첨자 : 1
# 커피 당첨자 : [2,3,4]
# == 축하합니다 ==

# (활용예제)
# from random import *
# lst = [1,2,3,4,5]
# print(lst)
# shuffle(lst) # 리스트의 순서가 무작위로 섞인다
# print(lst)
# print(sample(lst, 1)) # lst에서 1개를 무작위로 뽑는다

from random import *
num = range(1, 21) # 1부터 20까지 숫자를 생성
num = list(num) # 리스트로 변경
shuffle(num) # 리스트의 내용물 섞기
winner = sample(num, 4) # 치킨,커피 총 4명 뽑는다
print(winner)

print("== 당첨자 발표 ==")
print("치킨 당첨자 : [{}]".format(winner[0])) 
print("커피 당첨자 : {}".format(winner[1:]))
print("== 축하합니다 ==")
```

```python
#퀴즈 5
# 택시 기사, 50명의 승객과 매칭기회가 있다
# 총 탑승 승객 수를 구하는 프로그램

# 조건1 : 승객별 운행 시간은 5~50분 사이의 난수로
# 조건2 : 당신은 5~ 15분 사이의 승객만 매칭 해야함

# (예제문)
# [0] 1번쨰 손님 (소요시간 : 15분)
# [ ] 2번쨰 손님 (소요시간 : 50분)
# [0] 3번쨰 손님 (소요시간 : 5분)
# ...
# [ ] 50번째 손님 (소요시간 : 16분)
# 총 탑승 승객 : 2분

from random import *
total = 0
for i in range(1, 51): #50명 만들기
    time = randint(5, 50) #탑승시간 랜덤만들기
    if 5<= time <=15: #탑승시간이 조건2에 적합하면
        print("[0] {0}번쨰 손님 (소요시간 : {1}분)".format(i, time))
        total = total +1 #탑승을 시켰음으로 토탈값에 +1
    else: #매칠 실패한경우
        print("[ ] {0}번쨰 손님 (소요시간 : {1}분)".format(i, time))

print("총 탑승 승객: {}분".format(total))

```

```python
# 퀴즈6

# 표준체중 구하는 프로그램
# * 표준체중 : 각개인의 키에 적당한 체중

# 남자: 키*키*22
# 여자: 키*키*21

# 규칙1 : 표준체중은 별도의 함수 내에서 계산 
#         함수명: std_weight
#         전달값: 키 , 성별
# 규칙2 : 표준 체중은 소수점 둘째자리 표시

# (출력예제)
# 키 175cm 남자의 표준 체중은 67.38kg 입니다.

def std_weight(hight, sex):
    if sex == "남자":
        return hight*hight*22
    else:
        return hight*hight*21

hight = 175
sex = "남자"
weight = round(std_weight(hight/100, sex),2)
print("키 {0}cm {1}의 표준 체중은 {2}kg 입니다.".format(hight,sex,weight))
```

```python

#퀴즈7

# 회사에서 매주 1회 작성해야 하는 보고서 
# - x 주차 주간보고 -
# 부서 :
# 이름 :
# 업무 요약 :

# 1주부터 50주차 까지 보고서 파일을 만드는 프로그램

# 조건: 파일명 '1주차.txt' , '2주차.txt' 로 만든다

stat = 1
while True:
    with open("{0}주차.txt".format(stat),"w",encoding="utf8") as report:
        report.write("- {0} 주차 주간 보고 -".format(stat))
        report.write("\n부서 : ")
        report.write("\n이름 : ")
        report.write("\n업무 요약 : ")
        stat += 1
    if stat == 51:
        break

for i in range(1,51):
    with open(str(i)+"주차.txt","w",encoding="utf8") as report:
        report.write("- {0} 주차 주간 보고 -".format(i))
        report.write("\n부서 : ")
        report.write("\n이름 : ")
        report.write("\n업무 요약 : ")

```
